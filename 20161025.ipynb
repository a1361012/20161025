```python
import numpy as np
K = 9100
Interval = 500
Premium_Call = 179
Premium_Put = 185
ST = np.arange(K - Interval, K + Interval)
```

```python
#{math}.max() --> {}表函示庫 -->僅能對純量做運算
#{numpy}.maximum() --> 可向量運算
Payoff_LongCall = np.maximum(ST - K, 0) - Premium_Call  #Long Call
Payoff_ShortCall = - Payoff_LongCall                    #Short Call
Payoff_LongPut = np.maximum( K - ST, 0) - Premium_Put   #Long Put
Payoff_ShortPut = - Payoff_LongPut                      #Short Put
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#http://www.datahub.top/coursedetail/?id=29
#http://blog.sciencenet.cn/blog-251664-800766.html
#http://matplotlib.org/api/pyplot_summary.html matplotlib文件

plt.figure(1)
plt.subplot(221) #(211)
plt.ylim(-400,400) #Y軸
plt.title('LongCall') #標題
#plt.xlabel('K') #X軸標題
plt.plot(ST, Payoff_LongCall, linewidth = 2)
#plt.plot(ST, Payoff_ShortCall)
plt.axhline(0, color = 'r', ls= 'dashed')
plt.axhline(0 - Premium_Call, color = 'k', ls= 'dashed')
plt.axhline(0 + Premium_Put, color = 'k', ls= 'dashed')
plt.axvline(K, color = 'k', ls='--')
plt.axvline(K + Premium_Call, color = 'k', ls='--')


#plt.figure(1)
plt.subplot(222) #(212)
plt.ylim(-400,400)
plt.title('ShortCall')
#plt.plot(ST, Payoff_LongPut)
plt.plot(ST, Payoff_ShortCall, linewidth = 2)
plt.axhline(0, color = 'R', ls= 'dashed')
plt.axhline(0 - Premium_Call, color = 'K', ls= 'dashed')
plt.axhline(0 + Premium_Put, color = 'K', ls= 'dashed')
plt.axvline(K, color = 'k', ls='--')
plt.axvline(K + Premium_Call, color = 'k', ls='--')
plt.subplots_adjust(left=None, bottom=None, right= None, top=None, wspace=0.3, hspace=None) #修改間距
plt.figure(2)
plt.subplot(221)
plt.ylim(-400,400)
plt.title('LongPut')
plt.plot(ST, Payoff_LongPut, linewidth = 2)
#plt.plot(ST, Payoff_ShortPut)
plt.axhline(0, color = 'r', ls= 'dashed')
plt.axhline(0 - Premium_Call, color = 'k', ls= 'dashed')
plt.axhline(0 + Premium_Put, color = 'k', ls= 'dashed')
plt.axvline(K, color = 'k', ls='--')
plt.axvline(K - Premium_Put, color = 'k', ls='--')


plt.figure(2)
plt.subplot(222)
plt.ylim(-400,400)
plt.title('ShortPut')
#plt.plot(ST, Payoff_LongPut)
plt.plot(ST, Payoff_ShortPut, linewidth = 2)
plt.axhline(0, color = 'R', ls= 'dashed')
plt.axhline(0 - Premium_Call, color = 'K', ls= '--')
plt.axhline(0 + Premium_Put, color = 'K', ls= '--')
plt.axvline(K, color = 'k', ls='--')
plt.axvline(K - Premium_Put, color = 'k', ls='--')
plt.subplots_adjust(left=None, bottom=None, right= None, top=None, wspace=0.3, hspace=None)
plt.show()
```

```python
plt.figure(1) 
plt.ylim(-400,400) #Y軸
plt.title('Call')  #標題
#plt.xlabel('K')   #X軸標題
plt.plot(ST, Payoff_LongCall, linewidth = 2, label = 'Long')
plt.plot(ST, Payoff_ShortCall, linewidth = 2, label = 'Short')
plt.legend(bbox_to_anchor=(1.05, 1),loc = 2)
#plt.legend(loc = 'best') #新增圖例
plt.axhline(0, color = 'r', ls= 'dashed')
plt.axhline(0 - Premium_Call, color = 'k', ls= 'dashed')
plt.axhline(0 + Premium_Put, color = 'k', ls= 'dashed')
plt.axvline(K, color = 'k', ls='--')
plt.axvline(K + Premium_Call, color = 'k', ls='--')

plt.figure(2)
plt.ylim(-400,400)
plt.title('Put')
plt.plot(ST, Payoff_LongPut, linewidth = 2, label = 'long')
plt.plot(ST, Payoff_ShortPut, linewidth = 2, label = 'short')
#plt.legend(bbox_to_anchor=(1.05, 1),loc = 2) 圖一圖例寫法
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc = 3, ncol=2, mode="expand", borderaxespad=1.2) 
#圖例不同寫法
#bbox_to_anchor=(圖例左往右距離, 圖例下往上距離, 圖例寬度, unknown)
#ncol 圖例欄數 mode 圖例框樣式 borderaxespad 圖例到圖最上方之距離
plt.axhline(0, color = 'r', ls= 'dashed')
plt.axhline(0 - Premium_Call, color = 'k', ls= 'dashed')
plt.axhline(0 + Premium_Put, color = 'k', ls= 'dashed')
plt.axvline(K, color = 'k', ls='--')
plt.axvline(K - Premium_Put, color = 'k', ls='--')
```
